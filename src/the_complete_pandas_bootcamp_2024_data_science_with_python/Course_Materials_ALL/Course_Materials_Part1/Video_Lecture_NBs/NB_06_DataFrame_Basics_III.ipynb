### Ranking DataFrames with rank()

```python
import pandas as pd
```

```python
sales = pd.Series([15, 32, 45, 21, 55, 15, 0],  index = ["Mo", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"])
```

```python
sales = pd.Series([15, 32, 45, 15, 55, 15, 0],  index = ["Mo", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"])
```

```python
sales
```

```python
sales.sort_values(ascending = False)
```

```python
sales.rank(ascending=False, method = "min").sort_values(ascending = True)
```

```python
sales.rank(ascending=False, method = "min", pct=True).sort_values()
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.fare.rank(ascending = False)
```

```python
titanic["fare_rank"] = titanic.fare.rank(ascending = False, method="min")
```

```python
titanic.head()
```

```python
titanic.sort_values("fare", ascending= False)
```

```python
titanic.drop(columns = "fare_rank", inplace= True)
```

```python
titanic
```

### The agg() method

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.describe()
```

```python
#titanic.mean() # old
```

```python
titanic.mean(numeric_only=True) # new
```

```python
titanic.agg("mean") # old
```

```python
titanic.agg("mean", numeric_only=True) # new but not recommended
```

```python
titanic.select_dtypes("number").agg("mean") # new and best practise
```

```python
titanic.select_dtypes("number").agg(["mean", "std"])
```

```python
titanic.select_dtypes("number").agg(["mean", "std", "min", "max", "median"])
```

```python
titanic.select_dtypes("number").agg({"survived": "mean", "age":["min", "max"]})
```

### apply(), map() and applymap()

```python
import pandas as pd
```

```python
sales = pd.read_csv("sales.csv", index_col = 0)
```

```python
sales
```

```python
sales.info()
```

```python
sales.min(axis = 0)
```

```python
sales.min(axis = 1)
```

```python
def range_f(series):
    return series.max() - series.min()
```

```python
sales.apply(range_f, axis = 0)
```

```python
sales.apply(lambda x: x.max() - x.min(), axis = 0)
```

```python
summer = pd.read_csv("summer.csv")
```

```python
summer.head()
```

```python
summer.Athlete.apply(lambda x: x[0])
```

```python
summer.Athlete.map(lambda x: x[0])
```

```python
summer.iloc[:,1:3].applymap(lambda x: x[0]) #old
```

```python
summer.iloc[:,1:3].map(lambda x: x[0]) # new
```

```python
sales.applymap(lambda x: 0.4*x-5) # old
```

```python
sales.map(lambda x: 0.4*x-5) # new
```
